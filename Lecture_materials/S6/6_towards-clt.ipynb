## The birthday problem
What is the probability that among 50 people (at least) two have the same birthday? (Assuming all birthdays equally likely)

```python
import numpy as np

n_people = 50
n_mc = 10000

birthdays = np.random.randint(0, 365, size=(n_mc, n_people))

sorted_birthdays = np.sort(birthdays, axis=1)
has_collision = np.any(np.diff(sorted_birthdays) == 0, axis=1)
np.sum(has_collision) / n_mc
```

We can repeat the experiment with various numbers of people or samples to see how this probability changes.

## Tossing a coin
We toss a coin 100 times and count the number of heads. We repeat this experiment 100000 times and plot the histogram of the number of heads in each experiment. Interestingly, the histogram seems to have a bell shape!

```python
# (slow code; fast code example below)
import random

n_games = 100000
n_tosses = 100

tails_per_game = []
for i_game in range(n_games):
    n_tails = 0
    for i in range(n_tosses):
        toss = random.randint(0, 1)
        n_tails += 1 if toss == 0 else 0
    tails_per_game.append(n_tails)
        
sns.histplot(tails_per_game)
```

## Rolling a die
What is the distribution of the outcomes of rolling one die? It should have a uniform distribution if the die is not biased so each number has an equal probability of $\frac{1}{6}$.

```python
# Distribution of outcomes of rolling one die (in theory)

die_outcomes = np.arange(1, 6 + 1)
p_die = np.ones((6,)) / 6
plt.stem(die_outcomes, p_die)
plt.title("Die probabilities")
plt.xlabel("Number rolled")
plt.ylabel("Probability")
```

What is the distribution of the outcomes of summation of two dice?

```python
# Distribution of outcomes of summation of two dice

two_dice_table = die_outcomes[:, None] + die_outcomes[None, :]
two_dice_outcomes = np.unique(two_dice_table)
p_two_dice = [(two_dice_table == outcome).sum() / 6**2 for outcome in two_dice_outcomes]
plt.stem(two_dice_outcomes, p_two_dice)
plt.title("Sum of two dice")
plt.xlabel("Sum of numbers rolled")
plt.ylabel("Probability of sum")
```

What is the distribution of the outcomes of summation of three dice?

```python
# Distribution of outcomes of summation of three dice

three_dice_table = die_outcomes[:, None, None] + \
                   die_outcomes[None, :, None] + \
                   die_outcomes[None, None, :]
three_dice_outcomes = np.unique(three_dice_table)
p_three_dice = [(three_dice_table == outcome).sum() / 6**3 for outcome in three_dice_outcomes]
plt.stem(three_dice_outcomes, p_three_dice)
plt.title("Sum of three dice")
plt.xlabel("Sum of numbers rolled")
plt.ylabel("Probability of sum")
```
