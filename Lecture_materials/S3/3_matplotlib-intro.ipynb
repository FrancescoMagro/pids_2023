```python
dfdf.individuals = dfdf.individuals.astype(int)
dfdf = dfdf.sort_values(by="individuals")
dfdf.index = [1, 2, 3, 4, 5, 6, 0, 7]
dfdf = dfdf.sort_index()
dfdf.plot.barh(x="geomaster_name", y="individuals")
plt.barh(dfdf.geomaster_name, dfdf.individuals)
plt.savefig('ukraine1.pdf', dpi=300, bbox_inches='tight')
```

```python
dfdf
dfdf.index = [1, 2, 3, 4, 5, 6, 0, 7]
dfdf
```

```python
my_range = list(range(1, len(df.index) + 1))
fig, ax = plt.subplots(figsize=(6, 4))

ax.hlines(y=my_range, 
           xmin=0, 
           xmax=dfdf['individuals'] / 1e6, 
           color='#007ACC', 
           alpha=0.2, 
           linewidth=10
          )

ax.plot(dfdf['individuals'] / 1e6, 
         my_range, 
         'o', 
         markersize=10,
         markeredgewidth=0,
         color='#007ACC', 
         alpha=0.6
        )

ax.set_xlabel('Million refugees', 
              fontsize=18, 
              fontweight='black', 
              color = '#333F4B'
             )
ax.set_ylabel('')

ax.tick_params(axis='both', 
               which='major', 
               labelsize=12
              )

_ = plt.yticks(my_range,
           dfdf.geomaster_name
          )

# # add an horizonal label for the y axis 
fig.text(-0.27, 0.96, 
         'Receiving country', 
         fontsize=18, 
         fontweight='black', 
         color='#333F4B')

# # change the style of the axis spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.spines['left'].set_bounds((1, len(my_range)))
ax.set_xlim(0, 1.8)

ax.spines['left'].set_position(('outward', 8))
ax.spines['bottom'].set_position(('outward', 5))

plt.savefig('ukraine2.tiff', dpi=300, bbox_inches='tight')
```
