```python
x = np.linspace(-5, 5, 50)

f1 = np.sin(x)
f2 = np.cos(x)
f3 = 1 / (1 + x**2)
```

```python
plt.plot(x, f1, 'ro-')
plt.plot(x, f2, 'k*--')
plt.plot(x, f3, 'gv-.')

plt.legend(["A sine function", "A cosine function", "The Witch of Agnesi"])
```
